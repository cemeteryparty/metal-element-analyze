Loading Data from Google Drive

```python
# 運行後，點選網址，將授權碼輸至下方欄位，完成授權

import os
from google.colab import drive

file_list = []

drive.mount('/content/drive')
base_path = '/content/drive/My Drive/Colab Notebooks/期中project/Training/photos' # 圖片資料夾路徑，自行更改

for root, dirs, files in os.walk(base_path):
  for file in files:
    filename = os.path.join(root, file)
    file_list.append(filename)

file_list.sort()
```

```python
import numpy as np
from PIL import Image
import pandas as pd

# Image

img_list = []
size = (128, 128)

for i in range(len(file_list)):
  im = Image.open(file_list[i])
  im = im.resize(size, Image.BILINEAR)
  imarray = np.array(im)
  img_list.append(imarray)

x = np.asarray(img_list)

print(x.shape)

# Label

y = pd.read_csv('/content/drive/My Drive/Colab Notebooks/期中project/Training/label.csv', encoding = 'big5') # label檔案路徑，自行更改
y = y.drop(columns = 'file name')
```
